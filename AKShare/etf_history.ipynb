```python
# 北向净流入
jlr_north = ak.stock_hsgt_north_net_flow_in_em(symbol="北上")

# 获取2022年3月至今的净流入
jlr_north = jlr_north.loc[(jlr_north.date >= '2025-08-01') &
                          (jlr_north.date <= '2025-10-03')]

# 将value换算成亿元
jlr_north.value = round(jlr_north.value * 10000 / 100000000, 2)

# plot净流入柱状图
fig = px.bar(jlr_north, x="date", y="value", text="value",
             title="2022-03-01至2022-03-15北向资金净流入")
fig.update_traces(textposition="outside")

fig.update_xaxes(title="日期")
fig.update_yaxes(title="净流入（亿元）")

fig.show()
```
